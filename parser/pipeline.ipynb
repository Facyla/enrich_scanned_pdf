```python
from docling.document_converter import DocumentConverter
from docling.chunking import HybridChunker

conv_res = DocumentConverter().convert("https://arxiv.org/pdf/2206.01062")
doc = conv_res.document

chunker = HybridChunker(tokenizer="BAAI/bge-small-en-v1.5")  # set tokenizer as needed
chunk_iter = chunker.chunk(doc)
```

```python
for a in list(chunk_iter):
    print(a)
```

```python
import urllib.request
from io import BytesIO
from docling.backend.html_backend import HTMLDocumentBackend
from docling.datamodel.base_models import InputFormat
from docling.datamodel.document import InputDocument

url = "https://en.wikipedia.org/wiki/Duck"
text = urllib.request.urlopen(url).read()
in_doc = InputDocument(
    path_or_stream=BytesIO(text),
    format=InputFormat.HTML,
    backend=HTMLDocumentBackend,
    filename="duck.html",
)
backend = HTMLDocumentBackend(in_doc=in_doc, path_or_stream=BytesIO(text))
dl_doc = backend.convert()
print(dl_doc.export_to_markdown())
```

```python
from docling.document_converter import DocumentConverter
import os

def analyser_pdf(pdf_path):
    """
    Analyse un PDF et extrait son contenu avec docling
    
    Args:
        pdf_path (str): Chemin ou URL vers le PDF
    """
    # Initialiser le convertisseur
    converter = DocumentConverter()
    
    # Convertir le document
    result = converter.convert(pdf_path)
    
    # Obtenir le texte au format markdown
    markdown_content = result.document.export_to_markdown()
    
    # Obtenir les images et leur texte
    images = result.document.images
    for idx, image in enumerate(images):
        print(f"\n=== Image {idx + 1} ===")
        print("Texte extrait:", image.text())
        print("Métadonnées:", image.metadata)
    
    return markdown_content, images

# Utilisation
pdf_path = "https://arxiv.org/pdf/2408.09869"  # ou URL
markdown_content, images = analyser_pdf(pdf_path)

# Afficher le contenu markdown
print("\n=== Contenu du document ===")
print(markdown_content)
```
